```python
import requests
Link = "https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth"
link_text = requests.get(Link).text
print(link_text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(link_text,'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
print(soup.title)
print(soup.title.text)
```

```python
print(soup.a)
print(soup.find_all('a'))
```

```python
all_table  = soup.find_all('table')
print(all_table)
```

```python
our_table = soup.find('table',{'class':'wikitable sortable'})
print(our_table)
```

```python
table_links = our_table.find_all('a')
print(table_links)  
```

```python
#List of Indians by net worth
list_of = []
for links in table_links:
    list_of.append(links.get('title'))

print(list_of)
```

```python
import pandas as pd
df = pd.DataFrame(list_of)
print(df)
```
